```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Create a sample dataset
data = {
    'Name': ['Amit', 'Shami', 'Ankit', 'Dhiraj', 'Dhruvi', 'Payal', 'Prachi', 'Snehal', 'Urvashi', 'Priya','Rishi'],
    'Age': [25, 30, 28, 35, 22, 40, 45, 32, 27, 29,32],
    'Gender': ['Male', 'Male', 'Male', 'Male', 'Female', 'Female', 'Female', 'Female', 'Female', 'Female','Male'],
    'Occupation': ['Engineer', 'Data Scientist', 'Teacher', 'Doctor', 'Student', 'Manager', 'CEO', 'Designer', 'Software Developer','Artist','Marketing Specialist'],
    'Salary': [60000, 75000, 50000, 90000, 50000, 80000, 120000, 65000, 70000, 55000,8500 ]
}

df = pd.DataFrame(data)

# Display the DataFrame
df
```

```python
# Display basic information about the DataFrame
print("Basic Information about the DataFrame:")
print(df.info())

```

```python
# Display summary statistics of numerical columns
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Visualize the distribution of 'Age' using a histogram
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualize the distribution of 'Salary' using a boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(x='Salary', data=df, color='lightgreen')
plt.title('Distribution of Salary')
plt.xlabel('Salary')
plt.show()

```

```python
# Countplot for 'Gender'
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', data=df, palette='pastel')
plt.title('Count of Gender')
plt.show()
```

```python
 # Countplot for 'Occupation'
plt.figure(figsize=(8, 3))
sns.countplot(y='Occupation', data=df, palette='muted')
plt.title('Count of Occupation')
plt.xlabel('Count')
plt.show()
```

```python
# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Correlation matrix
correlation_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Scatter plot of Age vs. Salary
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Salary', data=df, hue='Gender', palette='Set2', s=100)
plt.title('Scatter Plot of Age vs. Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
# Average Salary by Occupation
avg_salary_by_occupation = df.groupby('Occupation')['Salary'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_salary_by_occupation.values, y=avg_salary_by_occupation.index, palette='viridis')
plt.title('Average Salary by Occupation')
plt.xlabel('Average Salary')
plt.ylabel('Occupation')
plt.show()
```

```python
# Pairplot for numerical variables
sns.pairplot(df, hue='Gender', palette='husl')
plt.suptitle('Pairplot of Numerical Variables', y=1.02)
plt.show()
```
